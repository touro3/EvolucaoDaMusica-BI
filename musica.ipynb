```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/Users/lucastourinho/musicaBI/dados/musicdata.csv')

filtered_data = data.dropna(subset=['value_actual'])

grouped_data = filtered_data.groupby(['year', 'format'])['value_actual'].sum().unstack()


plt.figure(figsize=(14, 7))
for column in grouped_data.columns:
    plt.plot(grouped_data.index, grouped_data[column], label=column)

plt.title('Evoluo das Vendas de Msica por Formato')
plt.xlabel('Ano')
plt.ylabel('Vendas (Unidades)')
plt.legend(title='Formato')
plt.grid(True)
plt.show()
```

```python
data = pd.read_csv('/Users/lucastourinho/musicaBI/dados/musicdata.csv')
filtered_data = data.dropna(subset=['value_actual'])

total_sales_by_format = filtered_data.groupby('format')['value_actual'].sum()

plt.figure(figsize=(14, 8))
total_sales_by_format.sort_values(ascending=False).plot(kind='bar', color='skyblue')

plt.title('Total de Vendas por Formato de Mdia')
plt.xlabel('Formato de Mdia')
plt.ylabel('Vendas Totais (Unidades)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  

plt.show()
```
